```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv('FIFA1_EDA.csv')
```

```python
df
```

```python
df.drop(['Crossing','Finishing','Heading Accuracy','Short Passing','Volleys','Dribbling','Curve','FK Accuracy','Long Passing','Ball Control','Acceleration','Sprint Speed','Agility','Reactions','Balance','Shot Power','Jumping','Stamina','Strength','Long Shots','Aggression','Interceptions','Positioning','Vision','Penalties','Composure','Marking','Standing Tackle','Sliding Tackle','GK Diving', 'GK Handling','GK Kicking', 'GK Positioning', 'GK Reflexes','day','month','dayofweek','Total',''],axis=1,inplace=True)
```

```python
df
```

```python
df.drop(['end_year','loan','empty','ID'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python

plt.figure(figsize=(10,8))
x=sns.countplot(data=df,x='Age')
x.set_xlabel(xlabel='Age of the players',fontsize=14)
x.set_title(label='Distribution of age of the players',fontsize=16)
plt.show()
```

```python
plt.figure(dpi=125)
sns.distplot(a=df['Age'],bins=20,kde=True,color='red')
plt.axvline(x=np.mean(df['Age']),c='green',ls='--',label='Mean age of all players')
plt.legend()
plt.xlabel('Age')
plt.ylabel('count')
plt.title('Distribution og age')
plt.show()
```
